**CONVENIO DE SUMA DE EINSTEIN**

Se denomina convenio de suma de Einstein, notación de Einstein o notación indexada a la convención utilizada para abreviar la escritura de sumatorios, en el que se suprime el símbolo de sumatorio representado con la letra griega sigma -
Σ\Sigma. El convenio fue introducido por Albert Einstein en 1916.1​ Se aplica en física en especial a los desarrollos realizados en Física avanzada. El convenio se aplica sólo a sumatorios sobre índices repetidos. El convenio se usa especialmente con tensores donde es muy frecuente la operación de suma sobre índices repetidos y sería muy fatigoso escribir explícitamente los signos de sumatorios.

```python
print('Einsum C =: \n')
print(np.einsum('ij,jk -> ik', A,B),'\n')
```

image.png

```python
A = np.array([[2,6,5,2],
              [2,-2,2,3],
              [1,5,4,0]])
B = np.array([[2,9,0,3],
              [3,6,8,-2],
              [1,3,5,0]])

print(f'Shape of A {A.shape} and shape of B {B.shape}')
#We need the same shape for Hardamond method
print('Hardamond C =: \n')
print(A*B, '\n')

print('Einsum C=: \n')
print(np.einsum('ij,ij -> ij',A,B))
```

```python
A = np.array([[2,6,5,2],
              [2,-2,2,3],
              [1,5,4,0]])
print('Transposed A =: \n')
print(A.T, '\n')

'''
I'm receiving a matrix which has different numbers in it's shape,
having at a start point a ij and a result of ji.

So we could define a transpose as a change in the shape of the matrix
'''
print('Einsum Transpose A =:')
print(np.einsum('ij -> ji',A),'\n')
```

```python

import tensorflow as tf### models
import pandas as pd ### reading and processing data
import seaborn as sns ### visualization
import numpy as np### math computations
import matplotlib.pyplot as plt### plotting bar chart
from tensorflow.keras.layers import Normalization, Dense, InputLayer
from tensorflow.keras.losses import MeanSquaredError, Huber, MeanAbsoluteError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
tensor_data = tf.constant(data)
tensor_data = tf.cast(tensor_data, tf.float32)
#If we use for example a float 16 we will get an infinte number, because it's too large to express it as a 16 byte
print(tensor_data)
```

```python
tensor_data = tf.random.shuffle(tensor_data)
print(tensor_data.shape)
print('---------------------------------------------')
print(f'This is tensor data in order to see the difference with tensor shuffle')
print(tensor_data [:5])
'''
In the tensor_data we can appreciate that the id's are in order
meanwhile in tensor_shuffle is not this is because, the function
tf.random.shuffle() makes a random order but following the data.

For example let's say we have [1,5,3],[3,2,5] and [8,23,35]
maybe the result is gonna be [3,2,5], [1,5,3] and [8,23,35], but what I'm trying
to say is that they will follow their data ---> I'm not understanding even
myself :)

'''
```

```python
#Remember that tensors are ordered as (y,x)

x = tensor_data[:,3:-1]
print(x.shape)
print(tensor_data.shape)
```

```python
y = tensor_data[:,-1]
print(y[:5].shape)
y = tf.expand_dims(y, axis = -1)
print(y[:5])
```

```python
normalizer = Normalization(axis = -1,mean = 5, variance = 4)
x_normalized =tf.constant([[3,4,5,6,7],
                           [4,5,6,7,8]])
normalizer(x_normalized)
```

```python
'''
x - mean / std

std**2 = variance

DESVIACIÓN TÍPICA = std
VARIAZA = std **2
MEDIA = mean

'''


#OBTAINING THE MEAN AND STANDARD DESVIATION THROUGH TENSORFLOW
normalizer = Normalization()
x_normalized =tf.constant([[3,4,5,6,7],
                           [4,10,6,7,8],
                           [32,1,56,3,5]])

normalizer.adapt(x_normalized)
normalizer(x_normalized)

```

```python
print(x.shape)
```

```python
TRAIN_RATIO = 0.8
VAL_RATIO = 0.1
TEST_RATIO = 0.1
DATASET_SIZE = len(x)
```

```python
x_train = x[:int(DATASET_SIZE * TRAIN_RATIO)]
y_train = y[:int(DATASET_SIZE * TRAIN_RATIO)]
print(x_train.shape,y_train.shape)
```

```python
x_val = x[int(DATASET_SIZE*TRAIN_RATIO):int(DATASET_SIZE*(TRAIN_RATIO+VAL_RATIO))]
y_val = y[int(DATASET_SIZE*TRAIN_RATIO):int(DATASET_SIZE*(TRAIN_RATIO+VAL_RATIO))]
print(x_val.shape)
print(y_val.shape)
```

```python

x_test = x[int(DATASET_SIZE*(TRAIN_RATIO+VAL_RATIO)):]
y_test = y[int(DATASET_SIZE*(TRAIN_RATIO+VAL_RATIO)):]
print(x_test.shape)
print(y_test.shape)
```

```python
normalizer = Normalization()
normalizer.adapt(x_train)
normalizer(x)[:5]
```

```python
model = tf.keras.Sequential([
                             InputLayer(input_shape = (8,)),
                             normalizer,
                             Dense(128, activation = "relu"),
                             Dense(128, activation = "relu"),
                             Dense(128, activation = "relu"),
                             Dense(1),
])

model.summary()
```

```python
tf.keras.utils.plot_model(model, to_file = 'model.png',show_shapes= True)
```

```python
from tensorflow.keras.losses import Huber, MeanAbsoluteError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python
#model.compile(loss = MeanSquaredError())
```

```python
model.compile(optimizer = Adam(learning_rate = 0.1),
              loss=MeanAbsoluteError(),
              metrics = RootMeanSquaredError())
```

```python
#model.compile(loss = Huber(delta = 0.2)) #This is a better way
```

```python
history = model.fit(x_train,y_train, validation_data=(x_val,y_val), epochs = 100, verbose = 1)
#The utility of verbose is to show us the data
```

```python
history.history
```

```python

plt.plot(history.history['root_mean_squared_error'])
plt.plot(history.history['val_root_mean_squared_error'])
plt.title('model performance')
plt.ylabel('rmse')
plt.xlabel('epoch')
plt.legend(['train', 'val'])
plt.show()
```
